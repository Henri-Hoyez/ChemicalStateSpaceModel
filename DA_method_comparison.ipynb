```python
import pandas as pd
import numpy as np 
import matplotlib
import matplotlib.pyplot as plt
```

```python
COTMIX_RESULT_PATH = "da_methods_results/results_cotmix.h5"
NAIVE_MODEL_PATH = "da_methods_results/naive_model.h5"
CLUDA_RESULT_PATH = "da_methods_results/cluda_results.h5"
RAINCOAT_PATH = "da_methods_results/results_raincoat.h5"
ADVSKM_PATH = "da_methods_results/results_advSKM.h5"
```

```python
df_cotmix = pd.read_hdf(COTMIX_RESULT_PATH)
df_cotmix = df_cotmix.drop(columns=["x_ticks"])

df_cotmix = df_cotmix/100

df_naive = pd.read_hdf(NAIVE_MODEL_PATH)

df_cluda = pd.read_hdf(CLUDA_RESULT_PATH)

df_raincoat = pd.read_hdf(RAINCOAT_PATH)
df_raincoat = df_raincoat/100

df_advskm = pd.read_hdf(ADVSKM_PATH)
df_advskm = df_advskm/100
```

```python
x_ticks= {
    "Amplitude shift": 5, 
    "Non Linear Surjective": 0, 
    "Non Linear Injective": 0, 
    "Output Noise": 0, 
    "Time Shift": 0, 
    "Causal Shift": 5, 
    "Frequency Shift": 0, 
    "Input Mean Shift": 3, 
    "Input Noise": 0,
    "Shift in Unit": 0, 
}

shift_names = [
    'Amplitude shift',
    'Non Linear Surjective',
    'Non Linear Injective',
    'Output Noise',
    'Time Shift',
    'Causal Shift',
    'Frequency Shift',
    'Input Mean Shift',
    'Input Noise',
    'Shift in Unit']

def to_string(array:np.ndarray):
    a = [f"{arr:0.2f}" for arr in array]
    return a

ampl_ticks = np.arange(0, 5, 0.5)-2.5
ampl_ticks = to_string(ampl_ticks)
ampl_ticks[5] = "0. (source)"
x_ticks['Amplitude shift'] = ampl_ticks
############

non_linear_surjective= np.arange(0, 1, 0.1)
non_linear_surjective = to_string(non_linear_surjective)
non_linear_surjective[0] = "0. (source)" 
x_ticks["Non Linear Surjective"] = non_linear_surjective
############

non_linear_injective = np.arange(1.1, 3.1, 0.2)
non_linear_injective = to_string(non_linear_injective)
non_linear_injective[0] = "0. (source)" 
x_ticks["Non Linear Injective"] = non_linear_injective
############

output_noise = np.arange(0.25, 5, 0.5)
output_noise = to_string(output_noise)
output_noise[0] = "0. (source)" 
x_ticks["Output Noise"] = output_noise
############

time_shift = np.arange(2., 22, 2)
time_shift = to_string(time_shift)
time_shift[0] = "0. (source)" 
x_ticks["Time Shift"] = time_shift
###########

causal_shift = np.arange(0., 1, 0.1)
causal_shift = to_string(causal_shift)
causal_shift[5] = "0.5\n (source)" 
x_ticks["Causal Shift"] = causal_shift
###########

frequency_shift = np.arange(3.5, 13.5, 1)
frequency_shift = to_string(frequency_shift)
frequency_shift[0] = "0. (source)" 
x_ticks["Frequency Shift"] = frequency_shift
###########

in_mean = np.arange(3, 13, 1)-2
in_mean = to_string(in_mean)
in_mean[2] = "0. (source)" 
x_ticks["Input Mean Shift"] = in_mean
###########

input_noise = np.arange(0.25, 5., 0.5)-0.25
input_noise = to_string(input_noise)
input_noise[0] = "0. (source)" 
x_ticks["Input Noise"] = input_noise
##########

shift_in_unit = np.arange(0, 1., .1)
shift_in_unit = to_string(shift_in_unit)
shift_in_unit[0] = "0. (source)" 
x_ticks["Shift in Unit"] = shift_in_unit
##########
```

```python
from matplotlib import colormaps

# from matplotlib import colormaps
# cmap = colormaps.get_cmap("tab10")



def compare_method_on_domain_shift(
        df_naive_model:pd.DataFrame, 
        df_methods:list, 
        method_names:list,
        ds_name:str, 
        figsize=(20, 8),
        title="What a Wonderful Title :)",
        x_ticks=None,
        x_label:str="Strength",
        ylims:tuple=(-0.01, 1.01),
        cmap=colormaps.get_cmap("tab10")):
    
    plt.figure(figsize=figsize)
    ax = plt.subplot(111)
    # plt.title(title)
    
    naive_means = df_naive_model.loc[ds_name].mean(axis=1).astype(float)
    naive_std = df_naive_model.loc[ds_name].std(axis=1).astype(float)

    naive_max = naive_means+ naive_std
    naive_min = naive_means- naive_std
    naive_max = np.clip(naive_max, 0, 1)

    ax.fill_between(naive_means.index.values, naive_max, naive_min, alpha=0.25, label='Naive accuracy std')
    naive_means.plot(ax=ax, label="Naive accuracy", style=".-")


    for i in range(len(df_methods)):
        color = cmap((i+1)/10)
        df_method = df_methods[i]
        method_name = method_names[i]

        try:
            method_performance = df_method.loc[ds_name]
        except KeyError:
            continue

        method_means = method_performance.mean(axis=1).astype(float)
        method_std = method_performance.std(axis=1).astype(float)
        method_min = method_means- method_std
        method_max = method_means+ method_std

        method_max = np.clip(method_max, 0, 1)

        ax.fill_between(method_means.index.values, method_max, method_min, alpha=0.25, label=f"{method_name} accuracy std", color=color)
        method_means.plot(ax=ax, label=f"{method_name} accuracy", style=".-", color=color)

    if not x_ticks is None:
        ax.set_xticks(range(1, len(x_ticks)+1))
        ax.set_xticklabels(x_ticks)

    plt.ylim(ylims)
    plt.xlabel(x_label, fontsize=24)
    plt.ylabel("Accuracy")
    plt.grid()
    plt.legend(bbox_to_anchor=(1.0, 1.022), loc='upper left', fontsize=18)
    
    plt.tight_layout()
    # plt.savefig(f'imgs/{ds_name}.pdf')
    plt.show()
```

```python
matplotlib.rcParams.update({'font.size': 18})

for k in shift_names:
    compare_method_on_domain_shift(
        df_naive_model=df_naive,
        df_methods=[df_cotmix, df_cluda, df_advskm, df_raincoat], 
        method_names=["CoTMix", "CLUDA", 'AdvSKM', "RAINCOAT"],
        ds_name=k, 
        x_ticks=x_ticks[k], 
        title=f"Performance Comparison on {k}",
        x_label=f"{k} Difference from Source",
        # x_label=f"$\\tau$",
        # x_label=f"$\\beta$",
        # x_label=f"Noise Variance $\\sigma$",
        figsize=(20, 8)
    )
```

```python
df_methods=[df_cotmix, df_cluda, df_advskm, df_raincoat]
method_names=["CoTMix", "CLUDA", 'AdvSKM', "RAINCOAT"]
noise_ticks = ['0.25\n (source)', '0.75', '1.25', '1.75', '2.25', '2.75', '3.25', '3.75', '4.25', '4.75']


# noise_ticks = ['0.25\n (source)', '', '', '1.75', '', '', '3.25', '', '', '4.75']

plt.figure(figsize=(20, 8))
ax = plt.subplot(121)

naive_means = df_naive.loc['Input Noise'].mean(axis=1).astype(float)
naive_std = df_naive.loc['Input Noise'].std(axis=1).astype(float)

naive_max = naive_means+ naive_std
naive_min = naive_means- naive_std
naive_max = np.clip(naive_max, 0, 1)

ax.fill_between(naive_means.index.values, naive_max, naive_min, alpha=0.25, label='Naive accuracy std')
naive_means.plot(ax=ax, label="Naive accuracy", style=".-")


for i in range(len(df_methods)):
    # color = cmap((i+1)/10)
    df_method = df_methods[i]
    method_name = method_names[i]

    method_performance = df_method.loc["Input Noise"]

    method_means = method_performance.mean(axis=1).astype(float)
    method_std = method_performance.std(axis=1).astype(float)
    method_min = method_means- method_std
    method_max = method_means+ method_std

    method_max = np.clip(method_max, 0, 1)

    ax.fill_between(method_means.index.values, method_max, method_min, alpha=0.25, label=f"{method_name} Accuracy std")
    method_means.plot(ax=ax, label=f"{method_name} accuracy", style=".-")

ax.set_xticks(range(1, len(noise_ticks)+1))
ax.set_xticklabels(noise_ticks)

ax.set_title("$(a)$",y=-.22)
ax.set_ylim(0., 1.0)
ax.set_xlabel("Noise Variance $\\sigma$")
ax.set_ylabel("Accuracy")
# ax.legend(loc=0, fontsize=12)
ax.grid(True)

ax = plt.subplot(122)

naive_means = df_naive.loc['Output Noise'].mean(axis=1).astype(float)
naive_std = df_naive.loc['Output Noise'].std(axis=1).astype(float)

naive_max = naive_means+ naive_std
naive_min = naive_means- naive_std
naive_max = np.clip(naive_max, 0, 1)

ax.fill_between(naive_means.index.values, naive_max, naive_min, alpha=0.25, label='Naive accuracy std')
naive_means.plot(ax=ax, label="Naive accuracy", style=".-")


for i in range(len(df_methods)):
    # color = cmap((i+1)/10)
    df_method = df_methods[i]
    method_name = method_names[i]

    method_performance = df_method.loc["Output Noise"]

    method_means = method_performance.mean(axis=1).astype(float)
    method_std = method_performance.std(axis=1).astype(float)
    method_min = method_means- method_std
    method_max = method_means+ method_std

    method_max = np.clip(method_max, 0, 1)

    ax.fill_between(method_means.index.values, method_max, method_min, alpha=0.25, label=f"{method_name} accuracy std")
    method_means.plot(ax=ax, label=f"{method_name} accuracy", style=".-")


ax.set_xticks(range(1, len(noise_ticks)+1))
ax.set_xticklabels(noise_ticks)

ax.legend(bbox_to_anchor=(1.0, 1.022), loc='upper left', fontsize=18)
plt.title("$(b)$",y=-.22)
ax.set_ylim(0., 1.0)
ax.set_xlabel("Noise Variance $\\sigma$")
# ax.set_ylabel("Accuracy")
ax.grid(True)

plt.tight_layout()
# plt.savefig(f'imgs/NoiseCombined.pdf')
plt.show()
```
